```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data.csv')
df.head()
```

```python
peso = df.loc[:,['Name','Weight']]
peso['Weight'] = peso['Weight'].str[0:3]
peso = peso.dropna()
peso['Weight'] = peso['Weight'].astype(int)
peso 
```

```python
a = len(peso.loc[(peso['Weight'] < 125)])
b = len(peso.loc[(peso['Weight'] >= 125) & (peso['Weight'] < 150)])
c = len(peso.loc[(peso['Weight'] >= 150) & (peso['Weight'] < 175)])
d = len(peso.loc[(peso['Weight'] >= 175)])
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = '< 125', '125 - 150', '150 - 175', '> 175'
sizes = [a, b, c, d]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
habilidad = df.loc[:,['Name','Overall']]
h1 = len(habilidad.loc[(habilidad['Overall'] == 40)])
h2 = len(habilidad.loc[(habilidad['Overall'] == 50)])
h3 = len(habilidad.loc[(habilidad['Overall'] == 60)])
h4 = len(habilidad.loc[(habilidad['Overall'] == 70)])
h5 = len(habilidad.loc[(habilidad['Overall'] == 80)])
h6 = len(habilidad.loc[(habilidad['Overall'] == 90)])
h7 = len(habilidad.loc[(habilidad['Overall'] == 100)])
```

```python
graf = [h1, h2, h3, h4, h5, h6, h7]
bars = ('40', '50', '60', '70', '80', '90', '100')
x_pos = np.arange(len(bars))

# Create bars with different colors
plt.bar(x_pos, graf, color=['black', 'red', 'green', 'blue', 'cyan', 'yellow', 'brown'])

# Create names on the x-axis
plt.xticks(x_pos, bars)

# Show graph
plt.show()
```
